```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib as plt
%matplotlib inline
import seaborn as sns
print('Setup Complete')
```

```python
calorie_filepath = '/kaggle/input/datasets/adilshamim8/predict-calorie-expenditure/test.csv'
calorie_data = pd.read_csv(calorie_filepath)
calorie_data.head()
```

```python
sns.lmplot(x='Duration', y='Heart_Rate', hue='Sex', data=calorie_data)
```

```python
sns.boxplot(x='Sex', y='Heart_Rate', data=calorie_data)
```

```python
selected_data = calorie_data[['Weight', 'Duration', 'Heart_Rate', 'Age']]
corr_data = selected_data.corr()
sns.heatmap(data=corr_data, annot=True)
```

## Conclusion

Through data visualization and correlation analysis, several significant insights were uncovered from this fitness dataset:

Firstly, there is a prominent positive correlation between **Duration** and **Heart_Rate**. As illustrated in the scatter plot with a regression line, heart rate rises steadily as the workout time increases. This observation is strongly backed by the correlation heatmap, where the correlation coefficient between `Duration` and `Heart_Rate` reaches a high value of **0.88**. This demonstrates that the length of time spent exercising is the primary driver of physical exertion and heart rate acceleration.

Secondly, interesting patterns emerged regarding biological factors like **Age** and **Weight** and **Sex**. According to the heatmap, the correlation coefficients between `Age` and `Heart_Rate` (0.017), as well as `Weight` and `Heart_Rate` (-0.0042), are extremely close to zero. This implies that within this specific dataset, an individual's age or body weight has negligible direct linear impact on their working heart rate. 

Lastly, the trend remains highly consistent across genders, as both male and female groups display nearly identical upward trajectories in heart rate over time.
